WAV:
WAV (Waveform Audio File Format)는 오디오 파일을 저장하기 위한 표준 파일 형식입니다. 이 형식은 1991년 마이크로소프트와 IBM에 의해 개발되었으며, 주로 Windows 시스템에서 사용됩니다.


WAV 파일 형식은 다음과 같은 주요 구조적 특징을 갖습니다:

헤더 정보: WAV 파일은 RIFF (Resource Interchange File Format) 기반으로, 파일의 시작 부분에 헤더 정보를 포함합니다. 이 헤더에는 파일 타입, 크기, 포맷 등의 정보가 담겨 있습니다.

포맷 청크 (Format Chunk): 이 부분은 오디오 데이터의 인코딩 방식을 정의합니다. 여기에는 오디오 채널의 수 (모노 또는 스테레오), 샘플링 빈도, 비트 깊이 (예: 16비트, 24비트) 등이 포함됩니다.

데이터 청크 (Data Chunk): 실제 오디오 데이터가 저장되는 부분입니다. 데이터는 일반적으로 PCM (Pulse Code Modulation) 형식으로 인코딩되어, 오디오의 원본 파형을 디지털 데이터로 변환합니다.

무손실 압축: WAV는 무손실 압축 방식을 사용하여, 오디오의 원본 품질을 그대로 유지합니다. 이는 파일 크기가 크지만, 오디오의 질이 높다는 장점을 가집니다.

멀티미디어 호환성: WAV 파일은 다양한 멀티미디어 플레이어와 음악 편집 소프트웨어에서 널리 호환되며, 특히 Windows 기반 시스템에서 많이 사용됩니다.

이러한 특성으로 인해, WAV 형식은 고품질 오디오를 필요로 하는 음악 제작 및 전문 오디오 편집 분야에서 선호되는 형식 중 하나입니다.


Pulse Code Modulation (PCM)은 아날로그 오디오 신호를 디지털 형식으로 변환하는 방법입니다. 이 과정은 세 단계로 구성됩니다: 샘플링, 양자화, 그리고 인코딩. 샘플링에서는 아날로그 신호를 특정 간격으로 측정하여 샘플을 생성합니다. 양자화 과정에서 각 샘플은 가장 가까운 표준값으로 변환되어 오디오의 다이내믹 레인지를 디지털 값으로 표현합니다. 마지막으로, 이러한 값들은 이진 숫자로 인코딩되어 디지털 오디오 데이터를 형성합니다. PCM은 무손실 압축 방식으로, 원본 아날로그 신호의 정확한 복제본을 제공합니다.

SPECTROGRAM:
스펙트로그램은 소리나 다른 신호의 주파수 스펙트럼이 시간에 따라 변하는 것을 시각적으로 표현한 것이다. 오디오 신호의 주파수 내용을 분석하는 데 주로 사용된다. 스펙트로그램에서 가로축은 시간을, 세로축은 주파수를 나타내며 색의 세기는 특정 시간에서의 특정 주파수의 세기를 나타낸다. 이 도구는 음성이나 환경 소음 등 다양한 소리의 특성을 분석하기 위해 신호 처리, 언어학, 음악 및 다양한 과학 분야에서 널리 사용된다. 스펙트로그램은 파형과 같은 시간 영역의 표현에서 쉽게 식별할 수 없는 패턴, 톤 및 소음을 식별하는 데 도움이 된다.
